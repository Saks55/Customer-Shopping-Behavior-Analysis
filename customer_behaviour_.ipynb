Which category is purchased the most?

```python
df['Category'].value_counts()

df['Category'].value_counts().plot(kind='bar')
plt.title("Most Purchased Category")
plt.show()
```

Top 10 purchased items?

```python
top_items = df['Item Purchased'].value_counts().head(10)
print(top_items)

top_items.plot(kind='bar')
plt.title("Top 10 Purchased Items")
plt.show()
```

Average purchase amount?

```python
df['Purchase Amount (USD)'].mean()
```

Maximum purchase amount?

```python
df['Purchase Amount (USD)'].max()
```

Minimum purchase amount?

```python
df['Purchase Amount (USD)'].min()
```

Which season has the highest shopping?

```python
df['Season'].value_counts()

df['Season'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Shopping by Season")
plt.show()
```

Most preferred payment method?

```python
df['Payment Method'].value_counts()

df['Payment Method'].value_counts().plot(kind='bar')
plt.title("Payment Method")
plt.show()
```

Top 10 shopping locations?

```python
top_location = df['Location'].value_counts().head(10)

print(top_location)

top_location.plot(kind='bar')
plt.title("Top Shopping Locations")
plt.show()
```

How many customers used discount?

```python
df['Discount Applied'].value_counts()

df['Discount Applied'].value_counts().plot(kind='bar')
plt.title("Discount Applied")
plt.show()
```

Subscription status of customers?

```python
df['Subscription Status'].value_counts()

df['Subscription Status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Subscription Status")
plt.show()
```
